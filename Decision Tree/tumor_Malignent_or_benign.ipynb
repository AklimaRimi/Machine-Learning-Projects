```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/ML projects/decision tree/data 2.csv'
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
data = data.drop('id',axis=1)
```

```python
data['diagnosis'].unique()
```

```python
data['Unnamed: 32'].unique()
```

```python
data = data.drop('Unnamed: 32',axis =1)
```

```python
data.shape
```

```python
label = data.iloc[:,0]
```

```python
feature = data.iloc[:,1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainF,testF,trainL,testL = train_test_split(feature,label,test_size = .3,random_state=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model =  DecisionTreeClassifier(criterion='gini',max_depth=10)
```

```python
model.fit(trainF,trainL)
```

```python
pred = model.predict(testF)
```

```python

from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(testL,pred))
```

```python
cm = confusion_matrix(testL,pred)
```

```python
import seaborn as plt
```

```python
plt.heatmap(cm,annot = True,fmt='d')
```
