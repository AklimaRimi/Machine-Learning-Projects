```python
data = pd.read_csv('/content/drive/MyDrive/ML projects/linear regression /CarPrice_Assignment.csv')

data = data.drop(['car_ID','symboling'],axis=1)

data.head()
```

```python
from word2number import w2n
data['doornumber'] = data['doornumber'].apply(w2n.word_to_num)
data['cylindernumber'] = data['cylindernumber'].apply(w2n.word_to_num)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['fueltype'] = le.fit_transform(data['fueltype'])
data['aspiration'] = le.fit_transform(data['aspiration'])
data['carbody'] = le.fit_transform(data['carbody'])
data['drivewheel'] = le.fit_transform(data['drivewheel'])
data['enginelocation'] = le.fit_transform(data['enginelocation'])
data['enginetype'] = le.fit_transform(data['enginetype'])
data['fuelsystem'] = le.fit_transform(data['fuelsystem'])

```

```python
data.head()
```

```python
feature = data.iloc[:,1:23]
label = data.iloc[:,23]
```

```python
from sklearn.model_selection import train_test_split

xfeature,yfeature,xlabel,ylabel = train_test_split(feature,label,test_size = .2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xfeature,xlabel)
```

```python
predict = model.predict(yfeature)
```

```python
j=0
for i in ylabel:
  
  print(i,round(predict[j],3))
  j = j+1
  
```
