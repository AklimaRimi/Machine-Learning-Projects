```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/ML projects/linear regression /Real estate.csv'
data =pd.read_csv(path)
data.head()
```

```python
data.shape
```

```python
feature = data.iloc[:,1:6]
label = data.iloc[:,7]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainF,testF,trainL,testL = train_test_split(feature,label,test_size=.3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(trainF,trainL)
```

```python
pred = model.predict(testF)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(testF['X2 house age'],testL,alpha=.4,color = 'red')
plt.scatter(testF['X2 house age'],pred,alpha=.3,color='blue')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(testF['X2 house age'],testF['X5 latitude'],pred, color='red')
ax.scatter3D(testF['X2 house age'],testF['X5 latitude'],testL, color='blue')
ax.view_init(-140, 160)
```
