```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/content/drive/MyDrive/ML projects/linear regression /insurance.csv'
```

```python
data = pd.read_csv(path)
data.head()
```

```python
from sklearn.preprocessing import  LabelEncoder
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
```

```python
data['smoker'] = le.fit_transform(data['smoker'])
data['region'] = le.fit_transform(data['region'])
```

```python
feature = data.iloc[:,0:5]
feature.head()
```

```python
label = data.iloc[:,6]
label.head()
```

```python
data.shape
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
model = LR()
model.fit(feature,label)
```

```python
from sklearn.model_selection import train_test_split
trainF,testF,trainL,testL = train_test_split(feature,label,test_size = .3)
```

```python
pred = model.predict(testF)
```

```python
pred
```

```python

plt.scatter(testF['age'],pred,alpha=.6,color='red')

plt.scatter(testF['age'],testL,color='black',alpha=.2)
#age	sex	bmi	children	smoker
plt.legend(['pred','actual'])


```
